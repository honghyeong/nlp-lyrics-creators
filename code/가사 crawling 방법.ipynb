```python
title=driver.find_elements_by_class_name('link_text')

title_song=[]
for i in title:
    title_song.append(i.text)
    
del title_song[0:4]
print(len(title_song))

#곡이름들은 link_text 라는 클래스에 저장되어 있어서, 빈 리스트를 생성하고 제목들을 append
#link_text 라는 클래스를 가진게 노래 제목 이전에 이상한게 좀 있어서 지워버리자
#사실 이부분은 총 곡이 몇갠지 알려고 만든부분..ㅎㅎ
```

```python
nums = list(range(1,90))
links = []
for num in nums:
    links.append(driver.find_element_by_xpath(f'//*[@id="content"]/div/div[3]/div[1]/div/\
    table/tbody/tr[{num}]/td[3]/div[1]/span/a').get_attribute('href'))
    
#이제 각 노래의 가사를 추출하려면 노래 링크를 찾아야 한다. 곡은 총 89곡이고, 각 곡의 absolute xpath 는
#중간의 {num}이라고 있는 부분에 곡의 숫자만 입력하면 된다.
#빈 리스트를 생성해서 링크를 다 저장했다!
```

```python
links
#잘 출력되는데, 앞에 주소는 필요없다!
```

```python
linknumbers = [link[29:] for link in links]
linknumbers
#그러니 날리자
```

```python
lyrics=[]
titles=[]
for linknumber in linknumbers:
    driver.get("https://vibe.naver.com/track/" + str(linknumber))
    time.sleep(1)
    title=driver.find_element_by_class_name("title")
    lyric=driver.find_element_by_class_name("lyrics")
    titles.append(title.text)
    lyrics.append(lyric.text)
    time.sleep(1)

#실제 가사를 크롤링해주는 부분! 곡 링크를 보면 앞부분은 같고 track/ 뒷부분의 고유코드만 다르기 때문에,
#링크넘버를 str로 변환해줘서 링크를 완성하고 .get함수로 연다
#제목과 가사는 title, lyrics 이란 클래스에 정의되어있기 때문에 그냥 text 부분을 빈 리스트에 append 해줬다.
#중간중간 너무 터져서 타임함수로 for문이 한번 돌때마다 2초동안 쉬게 해주자
#마지막에 다 안돌아가고 터지는데, 곡들중 inst 는 가사가 없어서 터지는거ㅠㅠ 하지만 다행히 바이브에선
#inst곡들은 맨 마지막에 두기때문에 터지면 딱 가사만 있는 곡의 제목과 가사를 긁어와!
```
